```python
from xml.etree.ElementTree import parse
```

```python
anno_paths = sorted(glob('BCCD_Dataset/BCCD/Annotations/*.xml'))
print(anno_paths)
```

```python
anno_path = anno_paths[0]
anno_path
```

```python
parser = parse(anno_path)
parser
```

```python
# folder に対応する箇所を検索
parser.find('folder')
```

```python
parser.find('folder').text
```

```python
parser.find('filename').text
```

```python
size = parser.find('size')
size
```

```python
size.find('width').text
```

```python
size.find('height').text
```

```python
size.find('depth').text
```

```python
objects = parser.findall('object')
objects
```

```python
obj = objects[0]
obj
```

```python
obj.find('name').text
```

```python
# 物体検出では背景 (BG: background) も 1 クラスと扱います。
bccd_labels = ['BG', 'RBC', 'WBC', 'Platelets']
```

```python
bccd_labels.index(obj.find('name').text)
```

```python
box = obj.find('bndbox')
```

```python
box.find('xmin').text
```

```python
box.find('ymin').text
```

```python
box.find('xmax').text
```

```python
box.find('ymax').text
```

```python
box = []
for tag in obj.find('bndbox'):
    val = int(tag.text)
    box.append(val)

box
```
